```python
!pip install pandas numpy scikit-learn nltk openpyxl -q

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import (
    euclidean_distances,
    manhattan_distances,
    cosine_similarity
)
from nltk.metrics import edit_distance
import nltk

nltk.download("punkt")
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx"

df = pd.read_excel(url)

print("Dataset Loaded")
print(df.shape)
```

```python
df = df.dropna(subset=["CustomerID"])
```

```python
customer = (
    df.groupby("CustomerID")
    .agg({
        "Quantity":"sum",
        "UnitPrice":"mean"
    })
    .reset_index()
)

customer.columns = [
    "CustomerID",
    "Total_Quantity",
    "Avg_Price"
]

print("\nCustomer Profiles")
print(customer.head())
```

```python
features = customer[
    ["Total_Quantity","Avg_Price"]
]

scaler = StandardScaler()

scaled = scaler.fit_transform(features)
```

```python
euclidean = euclidean_distances(scaled)

print("\nEuclidean Distance Matrix")
print(euclidean[:5,:5])
```

```python
manhattan = manhattan_distances(scaled)

print("\nManhattan Distance Matrix")
print(manhattan[:5,:5])
```

```python
cosine = cosine_similarity(scaled)

cosine_df = pd.DataFrame(
    cosine,
    index=customer["CustomerID"],
    columns=customer["CustomerID"]
)

print("\nCosine Similarity")
print(cosine_df.iloc[:5,:5])
```

```python
country1 = str(df["Country"].iloc[0])
country2 = str(df["Country"].iloc[50])

ed = edit_distance(
    country1,
    country2
)

print("\nEdit Distance")
print(country1, "-", country2)
print("Distance:", ed)
```

```python
cust = customer["CustomerID"].iloc[0]

similar = cosine_df.loc[cust]

# remove current customer
similar = similar.drop(cust)

# sort and select top 5
top5 = similar.sort_values(
    ascending=False
).head(5)

print("Selected Customer:", cust)

print("\nTop 5 Similar Customers:")
print(top5)
```
